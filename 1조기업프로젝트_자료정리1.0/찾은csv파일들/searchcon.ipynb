```python
import pandas as pd
```

```python
# 입장객 합계가 있는 엑셀 파일의 경로
touristall = "전국 관광지 연령대별 관광객수.xlsx"
# 관광지 정보가 있는 csv 파일의 경로
toursitedata = '전국관광지정보표준데이터.csv'
```

```python
# 입장객 합계 엑셀 파일 로드
df_touristall = pd.read_excel(touristall)

# 관광지 정보가 있는 csv 파일 로드
df_sitedata = pd.read_csv(toursitedata, encoding='cp949')
```

```python
from fuzzywuzzy import process

# 첫 번째 데이터프레임의 행 데이터와 두 번째 데이터프레임의 행 데이터를 비교하여 가장 유사한 행을 찾습니다.
threshold_similarity = 80  # 필요에 따라 유사도 임계값을 조정합니다.

selected_rows = []
for index, row1 in df_touristall.iterrows():
    name1 = row1['관광지']  # 첫 번째 데이터프레임에서 이름에 해당하는 컬럼을 지정해야 합니다.
    
    best_match = process.extractOne(name1, df_sitedata['관광지명'])[0]  # 두 번째 데이터프레임에서 이름에 해당하는 열 이름을 지정합니다.
    similarity_score = process.extractOne(name1, df_sitedata['관광지명'])[1]
    if similarity_score >= threshold_similarity:
        selected_rows.append((row1))

# 선택된 행 데이터로 새로운 데이터프레임을 생성합니다.
new_df = pd.DataFrame(selected_rows, columns=df_touristall.columns)  # 필요에 따라 열 이름을 수정합니다.
# new_df.dropna()
# 선택된 행 데이터를 출력합니다.
print(new_df)
```
